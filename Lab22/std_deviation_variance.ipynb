```python
import numpy as np
from matplotlib import pyplot as plt
```

# Standard Deviation And Variance

```python
data1 = [1,2,3,4,5]
data2 = [1,1,2,3,3]
data3 = [3,3,3]

print(f'data1 var: {np.var(data1)}')
print(f'data2 var: {np.var(data2)}')
print(f'data3 var: {np.var(data3)}')

print(f'data1 std: {np.std(data1)}')
print(f'data2 std: {np.std(data2)}')
print(f'data3 std: {np.std(data3)}')
```

```python
data1 = [1,2,3,4,5]
data2 = [1,1,2,3,3]
data3 = [3,3,3]

fig, ax = plt.subplots()

plt.axhline(y = np.mean(data1), color = 'r', linestyle = '--')
# plt.axhline(y = np.mean(data2), color = 'g', linestyle = '--')
# plt.axhline(y = np.mean(data3), color = 'b', linestyle = '--')

plt.scatter(np.arange(len(data1)),data1, c='r')
# plt.scatter(np.arange(len(data2)),data2, c='g')
# plt.scatter(np.arange(len(data3)),data3, c='b')


# Plot differences
data1_mean = np.mean(data1)

differences = [ el-data1_mean for el in data1]


for i, (x, diff) in enumerate(zip(data1, differences)):
    ax.plot([i, i], [data1_mean, x], 'g--', label='Difference from Mean' if i == 0 else '')
    ax.text(i, x, f'{diff:+.2f}', color='g', ha='center')

plt.legend()

```
