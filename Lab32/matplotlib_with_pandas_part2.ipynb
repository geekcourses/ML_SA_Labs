```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import colormaps
import numpy as np
import pandas as pd
```

```python
drinks_url = 'https://raw.githubusercontent.com/geekcourses/JupyterNotebooksExamples/master/datasets/various/drinks.csv'

df = pd.read_csv(drinks_url)
df.head()
```

```python
# scatterplot of beer servings versus wine servings
df.plot.scatter(x='beer_servings', y='wine_servings')
```

```python
# how np.where works
arr = np.arange(1,11)
print(arr)

# np.where(cond, A, B) works like: if(cond)? then A : else B, but on each element of the np.array

new_arr = np.where(arr%2==0,'Even', 'Odd')
print(new_arr)
```

```python
colors = np.where(df['country']=='France', 'cyan', 'black')
print(colors[:10])
```

```python
df.plot.scatter(x='beer_servings', y='wine_servings', color=colors)
```

```python
# color EU countries with 'blue', rest - black

colors_by_EU = np.where(df['continent']=='EU', 'blue', 'black')
df.plot.scatter(x='beer_servings', y='wine_servings', color=colors_by_EU)
```

```python
# np.full_like
# creat array of 10 values 'gray'
np.full_like(arr, 'gray', dtype='<U4')

# np.full_like(['abc', 'ddddd','q'], 'gray')
```

```python
# color EU countries with 'blue', AF with 'red', rest - black
colors = np.full_like(df['country'], 'black')
colors[df['continent']=='EU'] = 'blue'
colors[df['continent']=='AF'] = 'red'
colors
```

```python
df.plot.scatter(x='beer_servings', y='wine_servings', color=colors)
```

```python
colors = np.where(df['continent']=='EU', 'blue', 'black')
colors[df['continent']=='AF'] = 'red'
df.plot.scatter(x='beer_servings', y='wine_servings', color=colors)
```

```python
# TASK:
# EU countries with 'blue',
# AF with 'red',
# 'Bulgaria' with 'orange'
# rest - black

colors = np.where(df['continent']=='EU', 'blue', 'black')
colors = colors.astype('<U10')
colors[df['continent']=='AF'] = 'red'
colors[df['country']=='Bulgaria'] = 'orange'

df.plot.scatter(x='beer_servings', y='wine_servings', color=colors)
```

```python
# Calculate correlation matrix
corr = df.corr(numeric_only=True)
corr
```

```python
print( corr.columns)
print( corr.index)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Display the correlation matrix as an image
heatmap = ax.imshow(corr)

ax.set_xticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=45)

ax.set_yticks(np.arange(len(corr.columns)))
ax.set_yticklabels(corr.columns)

ax.set_title('Correlation Heatmap of Drinks Dataset')
plt.tight_layout()
```
