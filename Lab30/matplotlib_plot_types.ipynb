```python
import matplotlib.pyplot as plt
import numpy as np
```

## Scater

```python
# Sample Data (students' test scores)
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [22, 23, 24, 22, 20, 19, 21]

# Creating a scatter plot
plt.scatter(days, temperatures, c='blue', marker='o', label='Day Temp')
plt.title("Days vs. Temp ")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.legend()
plt.grid(True)
```

## Histogram

```python
# Sample data (test scores for a class of students)
scores = np.array([2,2,2,2,2,3,3,4,5,5,5,6,6,6,6,6,6],dtype="int")
print(scores.shape)

# Create the histogram
res = plt.hist(scores, bins=np.arange(1.5, 7.5), edgecolor='black')
print(res)

# Add labels and title
plt.xlabel('Test Score')
plt.ylabel('Number of Students')
plt.title('Distribution of Test Scores')
```

## Box Plot

```python
# Sample Data
job_types = ['Engineer', 'Manager', 'Analyst', 'Designer', 'Salesperson']
incomes = [60000, 65000, 75000, 80000, 95000, 1_000_000]

# print quartiles
print('quartiles', np.quantile(incomes,[0.25,0.5,0.75]))

# Creating a box plot with labels
fig, [ax1, ax2] = plt.subplots(nrows=2, ncols=1)

ax1.hist(incomes)
ax2.boxplot(incomes)
plt.title("Income Distribution by Job Type")
plt.ylabel("Income (USD)")
plt.xlabel('Job type')
plt.grid(True)

# plt.scatter(job_types, incomes)

plt.show()
```
