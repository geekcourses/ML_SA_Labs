```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
plt.subplot(121)
plt.plot(x, x**2, color="red")
plt.title('X**2')

plt.subplot(122)
plt.plot(x, x**3, color="green")
plt.title('X**3')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

axes_y = [x**2, x**3, x+2]
axes_titles = ['x**2', 'x**3', 'x+2']
axes_colors = ['r', 'g', 'b']

for idx in range(3):
    axes[idx].plot(x, axes_y[idx], label=axes_titles[idx], color=axes_colors[idx])
```
