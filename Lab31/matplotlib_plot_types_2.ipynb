```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data (total sales of a product)
products = ['toys', 'food', 'drinks']
sales = [100, 50, 200]

# Create the bar plot
plt.bar(products, height=sales, edgecolor='black', width=1, bottom=25, align='edge' )

# Add labels and title
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales")
```

```python
# Create the bar plot
plt.barh(products, width=sales, edgecolor='black', height=1, align='edge' )

# Add labels and title
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales")
```

```python
# Pie chart

# Sample data (total sales of a product)
products = ['toys', 'food', 'drinks']
sales = [100, 50, 200]
colors=['red', 'green','blue']
```

```python
plt.pie(x=sales, labels=products, colors=colors, autopct='%1.1f%%', pctdistance=0.9, shadow=True)
```
