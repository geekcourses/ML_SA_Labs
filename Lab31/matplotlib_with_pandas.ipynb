```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import pandas as pd
```

```python
# create a simple DataFrame
df = pd.DataFrame({'x': range(10), 'y': range(10, 20)})
df
```

```python
plt.scatter(df['x'], df['y'])
```

```python
df.plot(kind='scatter', x='x', y='y')
```

```python
df
```

```python
df.plot(kind='bar', x='x', title='Demo', grid=True)
```

```python
# create a simple DataFrame
df = pd.DataFrame({'y1': range(10, 20), 'y2': range(15, 25), 'y3': range(20, 30)})
df
```

```python
df.plot(kind='line', subplots=True, layout=(1,3), sharey=True)
```

```python
# plot 'y1' column
df.plot(y='y1')
```

```python
for column in df.columns:
    df.plot(y=column)
```
