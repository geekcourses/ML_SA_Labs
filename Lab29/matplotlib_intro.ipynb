```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11)
```

```python
fig = plt.figure()
fig.set_facecolor('lightgray')

# ax1 = fig.add_axes( (0,0,50,50) )

# add 2 plot, in 1 row*2cols
fig.add_subplot(121,facecolor='green', title='Figure 1')
fig.add_subplot(122, title='Figure 2')

plt.plot(x, x**2)
plt.plot(x, x**3)
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
fig.set_facecolor('lightgray')

ax1.plot(x, x**2)
ax2.plot(x, x**3)

ax1.set_facecolor('green')
ax1.set_title('Figure 1')
ax2.set_title('Figure 2')

ax1.set_xlabel('X')
ax1.set_ylabel('Y', rotation=0, loc='top')

x_ticks = ax1.get_xticklabels()
x_ticks[1].set_color('yellow')
```
